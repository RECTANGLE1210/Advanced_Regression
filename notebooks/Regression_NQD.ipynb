```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
traindf = pd.read_csv('../data/preprocessed/train_preprocessed.csv') #Đọc dữ liệu từ tập huấn luyện đã qua xử lí
traindf = traindf.rename(columns={'Unnamed: 0': 'Id'}) #Thêm tên cho cột Id
traindf
```

```python
features = traindf.drop(columns = ['Id', 'SalePrice']) #Bỏ 2 cột Id và SalePrice để lấy dataframe gồm những đặc điểm
labels = traindf['SalePrice'] #Lấy cột SalePrice để huấn luyện
features
```

```python
testdf = pd.read_csv('../data/preprocessed/test_preprocessed.csv') #Đọc dữ liệu từ tập kiểm tra đã qua xử lí
test = testdf.drop(columns = ['Id']) #Bỏ cột Id để kiểm tra
```

```python
raw = pd.read_csv('../data/raw/train.csv') #Đọc dữ liệu từ tập huấn luyện gốc
mean = raw['SalePrice'].mean() #Giá trị trung bình của SalePrice
std = raw['SalePrice'].std() #Độ lệch chuẩn của SalePrice
```

## Linear Regression

```python
#Huấn luyện mô hình với hồi quy tuyến tính
LinearRegr = LinearRegression()
LinearRegr.fit(features, labels)
```

```python
LinearRegr.predict(test) #Dự đoán kết quả, tuy nhiên đây không phải SalePrice thật sự
```

```python
#Tạo dataframe với hàm apply cho từng phần tử của Predict(test) thành SalePrice thật sự
SalePrice1 = pd.DataFrame(LinearRegr.predict(test), columns = ['SalePrice']).apply(lambda x: x*std + mean) 
```

```python
Id1 = pd.DataFrame(testdf['Id']) #Tạo dataframe cho Id
output = pd.concat([Id1, SalePrice1], axis = 1) #Kết hợp 2 dataframe
```

```python
#Lưu kết quả
output_dir = './data/output'
os.makedirs(output_dir, exist_ok=True)
output_filepath = os.path.join(output_dir, 'LinearRegr_predictions.csv')
output.to_csv(output_filepath, index=False)
```

```python
print(f"Predictions saved to {output_filepath}")
```

```python
#Huấn luyện mô hình với hồi quy Lasso
LassoRegr = Lasso(alpha = 0.001)
LassoRegr.fit(features, labels)
```
